```python
import numpy as np
from main.tokenizer_optimized import Tokenizer
```

```python
tok = Tokenizer.from_files(
    "trained_tokenizer/vocab.json",
    "trained_tokenizer/merges.json",
    special_tokens=["<|endoftext|>"]
)

files = {
    "train": "../bpe-optimized-from-scratch/data/TinyStoriesV2-GPT4-train.txt",
    "val":   "../bpe-optimized-from-scratch/data/TinyStoriesV2-GPT4-valid.txt",
}

for split, path in files.items():
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()
    ids = tok.encode(text)
    arr = np.array(ids, dtype=np.uint16)  # 你的 vocab_size=10000，用 uint16 没问题
    arr.tofile(f"tokenized_data/{split}.bin")
    print(split, "tokens:", len(arr), "saved ->", f"tokenized_data/{split}.bin")
```
